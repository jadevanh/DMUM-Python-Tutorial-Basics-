```python
df.columns = df.iloc[0]
df = df[1:].reset_index(drop=True)
```

```python
df.drop(columns=['Prefix'], inplace=False)
```

```python
# what data types are the columns?
# are they really all objects (refer to slides for meaning of object datatype)
df.dtypes
```

```python
# take a closer look at "Base Price"
# notice that this column is all numbers, but it is not considered a float, it is an object
df['Base Price']
```

```python
# making the type a float (decimal) in order to use functions on the donation amount
df['Base Price'] = df['Base Price'].astype(float)
df['Base Price'].dtype
```

```python
df.dtypes
```

```python
# use the .str.split() to create a list in a column, now we can separate on the types of donors
df['Donor Segments'] = df['Donor Segments'].str.split(',')
# creating a new df in order to separate the list into string
df[['All Donors', 'Donor Segments 2', 'Donor Segments 3']] = pd.DataFrame(df['Donor Segments'].to_list(), index=df.index)
# droping the original column
df = df.drop(columns=['Donor Segments'], inplace=False)
```

```python
# Replacing the None values to NaN values
df[['Donor Segments 2', 'Donor Segments 3']] = df[['Donor Segments 2', 'Donor Segments 3']].replace({None: np.nan})
```
